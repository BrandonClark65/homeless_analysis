```python
import pandas as pd
from database_connection import engine
import functools as ft
```

```python
# Function to add year column and remove year from column names
def remove_year(dataframe, year):
        # Add year column
        dataframe.insert(loc=0, column='Year', value=year)

        # Strip away year from column names
        dataframe.columns = dataframe.columns.str.rstrip(f', {year}')    
        
```

# 2007

```python
# Import csv into pandas dataframe 2007
data_2007 = pd.read_csv("Resources/2007-PIT-Counts-by-State.csv")
data_2007_df = pd.DataFrame(data_2007)
```

```python
# Run remove_year funtion
remove_year(data_2007_df, 2007)
# Display DataFrame
data_2007_df.head()
```

```python
# Export df to sql table
data_2007_df.to_sql(name='data_2007', con=engine, if_exists='replace')
```

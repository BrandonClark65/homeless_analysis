# 2007

```python
# Import csv into pandas dataframe
data_2007 = pd.read_csv("Resources/2007-PIT-Counts-by-State.csv")
data_2007_df = pd.DataFrame(data_2007)
data_2007_df.head()
```

```python
# Create column for the year and set to 2007
data_2007_df.insert(loc=0, column='Year', value=2007)
data_2007_df.head()
```

```python
# Strip away year from column names
data_2007_df.columns = data_2007_df.columns.str.rstrip(', 2007')
data_2007_df.head()
```

```python
# Export df to sql table
data_2007_df.to_sql(name='data_2007', con=engine, if_exists='replace')
```

```python
years = [2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
def remove_year():
    for year in years:
        # Import csv into pandas dataframe
        data = pd.read_csv(f"Resources/{year}-PIT-Counts-by-State.csv")
        data_df = pd.DataFrame(data)
        
        # Add year column
        data_df.insert(loc=0, column='Year', value=year)

        # Strip away year from column names
        data_df.columns = data_df.columns.str.rstrip(f', {year}')    
        
        print(f'Cleaning done for {year}')
```

```python
remove_year()
```
